```python
imagesdata = []


for index in indices: 
    url = (f"{base_url}&start={index}&api_key={api_key}")
    response = requests.get(url).json()

    data = response['data']

    for i in range(len(data)):
        
        images = data[i]['images']
        
        park = data[i]['fullName']
        
        images_df = pd.DataFrame(images)
        
        images_df['park'] = park
        
        imagesdata.append(images_df)
```

```python
all_images_df = pd.concat(imagesdata)
all_images_df.head(20)
```

```python
engine = sqlalchemy.create_engine('sqlite:///nps.sqlite', echo=False)


images_df.to_sql('all_images_df', con=engine, if_exists = "replace")
```
